```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/attacks.csv", encoding='latin1')
df = df.sample(n=200, random_state=42)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()

plt.figure()
missing.plot(kind='bar')
plt.title("Missing Values Before Cleaning")
plt.xticks(rotation=90)
plt.show()
```

```python
country_counts = df["Country"].value_counts()

plt.figure()
country_counts.plot(kind='bar')
plt.title("Top Countries (Before Cleaning)")
plt.show()
```

```python
# remove useless columns
df = df.drop(columns=["Unnamed: 22", "Unnamed: 23", "pdf", "href", "href formula"], errors='ignore')

# fix column names
df.columns = df.columns.str.strip()

# clean Fatal column
df["Fatal (Y/N)"] = df["Fatal (Y/N)"].str.strip()

# convert Age
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# drop duplicates
df = df.drop_duplicates()

# drop rows with missing Country/Year
df = df.dropna(subset=["Country", "Year"])
df.head()
```

```python
year_data = df["Year"].value_counts().sort_index()

plt.figure()
year_data.plot(kind='line')
plt.title("Attacks Over Years (After Cleaning)")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
fatal_counts = df["Fatal (Y/N)"].value_counts()

plt.figure()
fatal_counts.plot(kind='bar')
plt.title("Fatal vs Non-Fatal Attacks")
plt.show()
```

```python
plt.figure()
df["Age"].dropna().plot(kind='hist', bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
import plotly.express as px

country_data = df.groupby("Country").size().reset_index(name="attacks")

px.choropleth(country_data,
              locations="Country",
              locationmode="country names",
              color="attacks",
              title="Global Shark Attacks Distribution")
```

```python
activity_data = df.groupby(["Activity", "Fatal (Y/N)"]).size().reset_index(name="count")

fig = px.bar(activity_data,
             x="Activity",
             y="count",
             color="Fatal (Y/N)",
             animation_frame="Fatal (Y/N)",
             title="Activity vs Fatality")

fig.show()
```

```python
sunburst_data = df.groupby(["Country", "Activity"]).size().reset_index(name="count")

fig = px.treemap(sunburst_data,
                 path=["Country", "Activity"],
                 values="count",
                 color="count",
                 title="Shark Attacks Distribution")

fig.show()
```

```python
fig = px.violin(df,
                x="Fatal (Y/N)",
                y="Age",
                box=True,
                points="all",
                title="Age Distribution by Fatality")

fig.show()
```
